````python
## standart libraries
import os,sys
import numpy as np

# xarray
import xarray as xr

# for jupyter notebook display
%matplotlib inline
````

````python
diri='/Users/leroux/DATA/SASIP/PARAVIEW/data/'
month=1
nami = 'xSI_1982m0'+str(month)+'.nc'

diro=diri
````

````python
M = xr.open_dataset(diri+nami)
````

````python
timeyy = xr.DataArray(M.time.to_index().year.values,dims='time')
#timeyy= xrtime.broadcast_like(M.sst)
timeyy.name = 'timeyy'
timeyy.attrs['long_name'] = 'timeyy'
timeyy.attrs['standard_name'] = 'timeyy'

timemm = xr.DataArray(M.time.to_index().month.values,dims='time')
#timemm= xrtime.broadcast_like(M.sst)
timemm.name = 'timemm'
timemm.attrs['long_name'] = 'timemm'
timemm.attrs['standard_name'] = 'timemm'

timedd = xr.DataArray(M.time.to_index().day.values,dims='time')
#timedd= xrtime.broadcast_like(M.sst)
timedd.name = 'timedd'
timedd.attrs['long_name'] = 'timedd'
timedd.attrs['standard_name'] = 'timedd'

timehh = xr.DataArray(M.time.to_index().hour.values,dims='time')
#timehh= xrtime.broadcast_like(M.sst)
timehh.name = 'timehh'
timehh.attrs['long_name'] = 'timedd'
timehh.attrs['standard_name'] = 'timedd'
````

````python
timeyy.to_netcdf(diri+'xrtime_yymmddhh1d_'+str(month)+'.nc',mode='w')
timemm.to_netcdf(diri+'xrtime_yymmddhh1d_'+str(month)+'.nc',mode='a')
timedd.to_netcdf(diri+'xrtime_yymmddhh1d_'+str(month)+'.nc',mode='a')
timehh.to_netcdf(diri+'xrtime_yymmddhh1d_'+str(month)+'.nc',mode='a')
````
